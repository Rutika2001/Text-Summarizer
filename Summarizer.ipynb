```python
# Core Packages
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter.scrolledtext import *
import tkinter.filedialog


%run spacy_summarization.ipynb

# Web Scraping Pkg
from bs4 import BeautifulSoup
from urllib.request import urlopen

 # Structure and Layout
window = Tk()
window.title("Summarizer GUI")
window.geometry("850x500")
#window.config(background='black')

style = ttk.Style(window)

# TAB LAYOUT
tab_control = ttk.Notebook(window)

tab1 = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)
tab3 = ttk.Frame(tab_control)
tab4 = ttk.Frame(tab_control)


# ADD TABS TO NOTEBOOK
tab_control.add(tab1, text=f'{"Home ":^64s}')
tab_control.add(tab2, text=f'{"File":^64s}')
tab_control.add(tab3, text=f'{"URL":^64s}')
tab_control.add(tab4, text=f'{"Text":^64s}')

label1=Label(tab1,text="Text Summarizer",font=('impact',50,'bold'))
label1.place(x=180,y=180)    
    
label2 = Label(tab2, text= 'File Processing',padx=5, pady=5)
label2.grid(column=0, row=0)

label3 = Label(tab3, text= 'URL',padx=5, pady=5)
label3.grid(column=0, row=0)

tab_control.pack(expand=1, fill='both')


# Functions for tab4 text
def get_summary():
    raw_text = str(entry.get('1.0',tk.END))
    final_text = text_summarizer(raw_text)
    #print(final_text)
    result = '\nSummary:{}'.format(final_text)
    tab1_display.insert(tk.END,result)

# Clear entry widget
def clear_text():
    entry.delete('1.0',END)

def clear_display_result():
    tab1_display.delete('1.0',END)

#---


# Open File to Read and Process
def openfiles():
    file1 = tkinter.filedialog.askopenfilename(filetypes=(("Text Files",".txt"),("All files","*")))
    read_text = open(file1).read()
    displayed_file.insert(tk.END,read_text)


def get_file_summary():
    raw_text = displayed_file.get('1.0',tk.END)
    final_text = text_summarizer(raw_text)
    result = '\nSummary:{}'.format(final_text)
    tab2_display_text.insert(tk.END,result)
    

def clear_text_file():
    displayed_file.delete('1.0',END)   

def clear_text_result():
    tab2_display_text.delete('1.0',END)    

    
#---
# Fetch Text From Url
def get_text():
    raw_text = str(url_entry.get())
    page = urlopen(raw_text)
    soup = BeautifulSoup(page)
    fetched_text = ' '.join(map(lambda p:p.text,soup.find_all('p')))
    url_display.insert(tk.END,fetched_text)

def get_url_summary():
    raw_text = url_display.get('1.0',tk.END)
    final_text = text_summarizer(raw_text)
    result = '\nSummary:{}'.format(final_text)
    tab3_display_text.insert(tk.END,result)	
    
def clear_url_entry():
    url_entry.delete(0,END)

def clear_url_display():
    tab3_display_text.delete('1.0',END)


#--------------------------------------------------------------------------------------------------------------------------

#FILE PROCESSING TAB
l1=Label(tab2,text="Open File To Summarize")
l1.grid(row=1,column=1)

displayed_file = ScrolledText(tab2,height=10)
displayed_file.grid(row=2,column=0, columnspan=3,padx=100,pady=3)

# BUTTONS FILE READING TAB
b0=Button(tab2,text="Open File", width=12,command=openfiles,bg='#03A9F4',fg='black')
b0.grid(row=3,column=0,padx=10,pady=10)

b1=Button(tab2,text="Reset ", width=12,command=clear_text_file,bg='#03A9F4',fg='black')
b1.grid(row=3,column=1,padx=10,pady=10)

b2=Button(tab2,text="Summarize", width=12,command=get_file_summary,bg='#03A9F4',fg='black')
b2.grid(row=3,column=2,padx=10,pady=10)

b3=Button(tab2,text="Clear Result", width=12,command=clear_text_result,bg='#03A9F4',fg='black')
b3.grid(row=7,column=1,padx=10,pady=10)

b4=Button(tab2,text="Close", width=12,command=window.destroy,bg='#03A9F4',fg='black')
b4.grid(row=7,column=2,padx=10,pady=10)

# Display Screen
tab2_display_text = ScrolledText(tab2,height=9)
tab2_display_text.grid(row=5,column=0, columnspan=3,padx=100,pady=5)

# Allows  to edit
tab2_display_text.config(state=NORMAL)

#-----------------------------------------------------------------------------------------------------------------------

# URL TAB
l1=Label(tab3,text="Enter URL To Summarize")
l1.grid(row=1,column=0)

raw_entry=StringVar()
url_entry=Entry(tab3,textvariable=raw_entry,width=50)
url_entry.grid(row=1,column=1)

# BUTTONS
button1=Button(tab3,text="Reset",command=clear_url_entry, width=12,bg='#03A9F4',fg='black')
button1.grid(row=4,column=0,padx=100,pady=10)

button2=Button(tab3,text="Get Text",command=get_text, width=12,bg='#03A9F4',fg='black')
button2.grid(row=4,column=1,padx=1,pady=10)


button3=Button(tab3,text="Clear Result", command=clear_url_display,width=12,bg='#03A9F4',fg='black')
button3.grid(row=7,column=0,padx=100,pady=10)

button4=Button(tab3,text="Summarize",command=get_url_summary, width=12,bg='#03A9F4',fg='black')
button4.grid(row=7,column=1,padx=10,pady=10)

# Display Screen For Result
url_display = ScrolledText(tab3,height=9)
url_display.grid(row=5,column=0, columnspan=3,padx=50,pady=5)


tab3_display_text = ScrolledText(tab3,height=9)
tab3_display_text.grid(row=10,column=0, columnspan=3,padx=100,pady=5)

#-----------------------------------------------------------------------------------------------------------------------
#TAB4
l1=Label(tab4,text="Enter Text To Summarize")
l1.grid(row=1,column=0)

entry=Text(tab4,height=10)
entry.grid(row=2,column=0,columnspan=2,padx=100,pady=5)

# BUTTONS
button1=Button(tab4,text="Reset",command=clear_text, width=12,bg='#03A9F4',fg='black')
button1.grid(row=4,column=0,padx=10,pady=10)

button2=Button(tab4,text="Summarize",command=get_summary, width=12,bg='#03A9F4',fg='black')
button2.grid(row=4,column=1,padx=10,pady=10)

button3=Button(tab4,text="Clear Result", command=clear_display_result,width=12,bg='#03A9F4',fg='black')
button3.grid(row=10,column=0,padx=10,pady=10)

button4=Button(tab4,text="Close", width=12,bg='#03A9F4',fg='black',command=window.destroy)
button4.grid(row=10,column=1,padx=10,pady=10)

# Display Screen For Result
tab1_display = ScrolledText(tab4,height=10)
tab1_display.grid(row=5,column=0, columnspan=3,padx=100,pady=5)



window.mainloop()

```
